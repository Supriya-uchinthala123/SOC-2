```python
#rename the sex field in titanic dataset to "Gender"
names={'Sex':"Gender"}
df_titanic=df_titanic.rename(columns=names)
```

```python
df_titanic
```

```python
#convert the male to 0 and female to 1 in titanic dataset
df_titanic.replace(to_replace="male",value=0,inplace=True)
df_titanic.replace(to_replace="female",value=1,inplace=True)
```

```python
df_titanic
```

```python
#select the data who are age less than 25 and gender is 1
df_titanic.loc[(df_titanic['Age']<25)&(df_titanic['Gender']==1)]
```

```python
#check male how many have survived and female how many have survived
female=len(df_titanic['Survived'])-(df_titanic['Gender']==0).sum()
male=len(df_titanic['Survived'])-(df_titanic['Gender']==1).sum()
print(female)
print(male)
```

```python
#people dead
len(df_titanic)-df_titanic['Survived'].sum()
```

```python
#plot a graph to show the count of survivors and non-survivors for each embark plot
survivors=df_titanic['Survived'].sum()
print(survivors)
non_survivors=len(df_titanic)-df_titanic['Survived'].sum()
print(non_survivors)
counts=[survivors,non_survivors]
names=['survivors','non_survivors']
plt.bar(names,counts,width=0.3)
```
